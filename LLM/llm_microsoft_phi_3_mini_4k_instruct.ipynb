# Hands-On Large Language Models:

## HuggingFace "microsoft/Phi-3-mini-4k-instruct" Model

The Phi-3-Mini-4K-Instruct is a 3.8B parameters, lightweight, state-of-the-art open model trained with the Phi-3 datasets that includes both synthetic data and the filtered publicly available websites data with a focus on high-quality and reasoning dense properties.

```python
# !pip install transformers
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained('microsoft/Phi-3-mini-4k-instruct')
```

```python
from transformers import AutoConfig, AutoModelForCausalLM

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    torch_dtype="auto",
    trust_remote_code=True,
)
```

```python
from transformers import pipeline

# Create a pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,    # this helps not to return the prompt but merely the output of the model
    max_new_tokens=500,        # This is the maximum number of tokens the model will generate. This helps not to continuly generate output until they reach their context window.
    do_sample=False            # The model will always select the next most probable token. This is for whether to use sampling for the next tokens to be generate!
)
```

```python
# The prompt (user input / query)
messages = [
    {"role": "user", "content": "What are large language models"}
]

# Generate output
output = generator(messages)
print(output[0]["generated_text"])
```
