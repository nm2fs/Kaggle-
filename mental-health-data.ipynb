## Handling outliers:

**Outliers can be detected using statistical methods like:**

**Z-score:** Measures the number of standard deviations a data point is from the mean.

**IQR (Interquartile Range)**: Identifies outliers based on the spread of the middle 50% of data.

```python
# Function to detect outliers using IQR for multiple columns
def detect_outliers_iqr_multi(df_train, columns):
    outlier_rows = pd.DataFrame(index=df_train.index)  # Initialize a DataFrame for tracking outliers
    
    for column in columns:
        Q1 = df_train[column].quantile(0.25)
        Q3 = df_train[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Identify outliers for this column
        outliers = (df_train[column] < lower_bound) | (df_train[column] > upper_bound)
        outlier_rows[column] = outliers  # Store outlier flags for this column
    
    # Combine outlier flags for all columns (True if any column has an outlier)
    outlier_rows['Is_Outlier'] = outlier_rows.any(axis=1)
    print("\nOutlier detection results (True indicates an outlier):\n", outlier_rows)
    
    # Filter rows with any outliers
    outlier_indices = outlier_rows[outlier_rows['Is_Outlier']].index
    return df_train.loc[outlier_indices]

# Detect outliers
columns_to_check = [
    'Age', 
    'Academic Pressure', 
    'Work Pressure', 
    'CGPA', 
    'Study Satisfaction', 
    'Job Satisfaction', 
    'Work/Study Hours', 
    'Financial Stress'
]
outliers = detect_outliers_iqr_multi(df_train, columns_to_check)
print("\nRows with outliers:\n", outliers)

# Remove rows with outliers
def remove_outliers_iqr_multi(df_train, columns):
    for column in columns:
        Q1 = df_train[column].quantile(0.25)
        Q3 = df_train[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Filter rows to keep only non-outliers
        df_train = df_train[(df_train[column] >= lower_bound) & (df_train[column] <= upper_bound)]
    return df_train

# Remove outliers
df_train_no_outliers = remove_outliers_iqr_multi(df_train, columns_to_check)
print("\nDataFrame after removing outliers:\n", df_train_no_outliers)
```

### I am using HistGradientBoostingClassifier because there are a lot of NaN (missing data) in the datasets and HistGradientBoostingClassifier could handle the missing data well without imputation.

```python
# # Handle missing values
# df_train = df_train.fillna(df_train.median())
# df_test = df_test.fillna(df_train.median())

# Split features (X) and target (y) from the train dataset
X_train = df_train.drop(columns=['Depression'], errors='ignore')
y_train = df_train['Depression']

# Standardize numerical columns for better performance
numerical_columns = X_train.select_dtypes(include=['int64', 'float64']).columns
scaler = StandardScaler()
X_train[numerical_columns] = scaler.fit_transform(X_train[numerical_columns])
df_test[numerical_columns] = scaler.transform(df_test[numerical_columns])

from sklearn.ensemble import HistGradientBoostingClassifier


# Train the model
model = HistGradientBoostingClassifier(random_state=42)
model.fit(X_train, y_train)

# Predict on df_test
df_test['Predicted_Depression'] = model.predict(df_test)
```

The HistGradientBoostingClassifier in sklearn is an efficient, histogram-based implementation of gradient boosting designed for large datasets. It is particularly useful when dealing with tabular data and can handle both categorical and numerical features.

If you want to use it specifically with ROC and AUC evaluations, here's how it works:

1. Why HistGradientBoostingClassifier?
* Efficiency: Faster training due to histogram-based techniques.
* Handles Missing Data: Supports missing values directly without imputation.
* Supports Categorical Features: Allows for better handling of categorical features by setting the categorical_features parameter.
* Scalability: Performs well on larger datasets.

## Since we don't have access to test true labels, we use cross validation

Cross-validation is a reliable alternative to evaluate the model's performance on the training dataset without needing the test labels.
Cross-validation divides the training data into multiple "folds" (e.g., 5 folds if cv=5) and performs the following steps:

* Splits the data into training and validation subsets for each fold.
* Trains the model on the training subset.
* Evaluates the model's performance on the validation subset.
* Repeats this process across all folds to ensure the evaluation is robust and not overly dependent on a single split of the data.
* This process provides:

An unbiased estimate of the model's performance because it tests on unseen validation data in each fold.
More reliable metrics, as the performance is averaged across folds, reducing the risk of overfitting to a single training/validation split.

```python
from sklearn.model_selection import cross_val_score

# Evaluate the model using cross-validation
cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')

print("Cross-Validation Accuracy Scores:", cv_scores)    # Scores for 5 folds
print("Mean Accuracy:", cv_scores.mean())  # The average accuracy across all folds, providing a single summary metric for how well the model is expected to generalize to unseen data.
```

```python
from sklearn.model_selection import cross_val_score

# Calculate F1-score
## Useful when there’s an imbalance in the dataset 

f1_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='f1')
print("F1 Scores:", f1_scores)
print("Mean F1 Score:", f1_scores.mean())

# Calculate ROC AUC
## Evaluates the model’s ability to distinguish between classes across different thresholds.

roc_auc_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc')
print("ROC AUC Scores:", roc_auc_scores)
print("Mean ROC AUC Score:", roc_auc_scores.mean())

# Calculate Precision
## Useful when the cost of false positives is high.

precision_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='precision')
print("Precision Scores:", precision_scores)
print("Mean Precision Score:", precision_scores.mean())
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# Get predicted probabilities for the positive class
y_pred_prob = model.predict_proba(X_train)[:, 1]  # Probability for class 1 (positive class)

# Compute ROC curve and AUC score
fpr, tpr, thresholds = roc_curve(y_train, y_pred_prob)
auc_score = roc_auc_score(y_train, y_pred_prob)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {auc_score:.2f})")
plt.plot([0, 1], [0, 1], color="gray", linestyle="--", label="Random Guessing")
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```
